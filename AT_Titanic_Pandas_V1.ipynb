```python
# Load the dataset
import pandas as pd

titanic_df = pd.read_csv('train.csv')

# Display the first few rows
titanic_df.head()

```

```python
# Data overview - info and describe
titanic_df.info()
titanic_df.describe()
```

```python
# Check for missing values
titanic_df.isnull().sum()
```

```python
# Handle missing values - fill Age with median
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())

# Fill Embarked with most common value
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')

# Check missing values after imputation
titanic_df.isnull().sum()
```

```python
# Data exploration - distribution of values
titanic_df['Sex'].value_counts()
titanic_df['Pclass'].value_counts()
titanic_df['Embarked'].value_counts()
```

```python
# Calculate survival rates by sex, class, and embarkation point
titanic_df.groupby("Sex")["Survived"].mean()
titanic_df.groupby("Pclass")["Survived"].mean()
titanic_df.groupby("Embarked")["Survived"].mean()
```

```python
# Additional exploration - create family size feature
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
titanic_df['FamilySize'].value_counts()
```

```python
# Survival rate by family size
titanic_df.groupby('FamilySize')['Survived'].mean()
```

```python
# Look at correlation between features and survival
titanic_df.corr(numeric_only=True)['Survived'].sort_values(ascending=False)
```

```python
# Group survival by class and sex together
titanic_df.groupby(['Pclass', 'Sex'])['Survived'].mean()
```

```python
# Save cleaned DataFrame for later modelling
titanic_df.to_csv("AT_Titanic_Cleaned_V1.csv", index=False)
print("Saved AT_Titanic_Cleaned_V1.csv")
```
